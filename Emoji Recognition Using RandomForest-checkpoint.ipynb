```python
flattensize_t = size_t[1]*size_t[0]
flattensize_y = size_y[1]*size_y[0]
flattensize_d = size_d[1]*size_d[0]
```

```python
for i in range(len_t):
    tongue_gray[i] = np.ndarray.flatten(tongue_gray[i]).reshape(flattensize_t,1)
```

```python
for i in range(len_y):
    yum_gray[i] = np.ndarray.flatten(yum_gray[i]).reshape(flattensize_y,1)
```

```python
for i in range(10):
    disgust_gray[i] = np.ndarray.flatten(disgust_gray[i]).reshape(flattensize_d,1)
```

```python
tongue_gray = np.dstack(tongue_gray)
yum_gray = np.dstack(yum_gray)
disgust_gray = np.dstack(disgust_gray)
```

```python
tongue_gray.shape
```

```python
tongue_gray=np.rollaxis(tongue_gray,axis=2,start=0)
yum_gray=np.rollaxis(yum_gray,axis=2,start=0)
disgust_gray=np.rollaxis(disgust_gray,axis=2,start=0)
```

```python
tongue_gray=tongue_gray.reshape(len_t,flattensize_t)
yum_gray=yum_gray.reshape(len_y,flattensize_y)
disgust_gray=disgust_gray.reshape(len_d,flattensize_d)
```

```python
tongue_gray.shape
```

```python
tongue_data = pd.DataFrame(tongue_gray)
tongue_data['label'] = "Face With Tongue"
yum_data = pd.DataFrame(yum_gray)
yum_data['label'] = "Smiling"
disgust_data = pd.DataFrame(disgust_gray)
disgust_data['label'] = "Disgusting"
```

```python
data1 = pd.concat([tongue_data,yum_data])
data = pd.concat([data1,disgust_data])
```

```python
from sklearn.utils import shuffle
```

```python
data_index = shuffle(data).reset_index()
```

```python
emoji_data = data_index.drop(['index'],axis=1)
```

```python
emoji_data.to_csv("Emoji.csv")
```

```python
x = emoji_data.values[:,:-1]
y = emoji_data.values[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import decomposition
```

```python
pca = decomposition.PCA(n_components=20, whiten=True, random_state=1)
```

```python
pca.fit(x_train)
```

```python
x_train_pca = pca.transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
eigen = (np.reshape(x[10],(512,512)).astype(np.float64))
```

```python
eigen
```

```python
fig = plt.figure(figsize=(30,30))
for i in range(10):
    ax = fig.add_subplot(2, 5, i+1, xticks=[], yticks=[])
    ax.imshow(pca.components_[i].reshape(eigen.shape), cmap=plt.cm.bone)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import classification_report, accuracy_score
```

```python
RF = RandomForestClassifier()
RF.fit(x_train_pca, y_train)
```

```python
y_pred=RF.predict(x_test_pca)
```

```python
y_pred
```

```python
for i in (np.random.randint(0,6,6)):
    predicted_image=(np.reshape(x_test[i],(512,512)).astype(np.float64))
    plt.title('predicted lable:{0}'.format(y_pred[i]))
    plt.imshow(predicted_image, interpolation='nearest',cmap='gray')
    plt.show()
```

```python
RF = RandomForestClassifier(n_estimators=100, random_state=42)
RF.fit(x_train_pca, y_train)
y_pred = RF.predict(x_test_pca)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy*100:.2f}%")
```

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
